```python
import pandas as pd

# Creating a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}

df = pd.DataFrame(data)

# Adding a new column using inplace=True
df['City'] = ['New York', 'San Francisco', 'Los Angeles']

# Displaying the DataFrame after the addition of the new column
print("DataFrame after adding 'City' column:")
print(df)

# Dropping a column using inplace=True
df.drop('Age', axis=1, inplace=True)

# Displaying the DataFrame after dropping the 'Age' column
print("\nDataFrame after dropping 'Age' column:")
print(df)
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}

df = pd.DataFrame(data)

# Adding a new column using inplace=True
df['City'] = ['New York', 'San Francisco', 'Los Angeles']

# Displaying the DataFrame after the addition of the new column
print("DataFrame after adding 'City' column:")
print(df)

# Dropping a column using inplace=True
df.drop('Age', axis=1, inplace=False)

# Displaying the DataFrame after dropping the 'Age' column
print("\nDataFrame after dropping 'Age' column:")
print(df)
```
